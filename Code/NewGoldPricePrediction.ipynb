```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import pickle

# Load and preprocess the data
df = pd.read_csv('../Database/gold_rate.csv')

# Convert the date column to a numeric feature
df['date_num'] = (pd.DatetimeIndex(df['Date']).year % 100) * 10000 + \
                 pd.DatetimeIndex(df['Date']).month * 100 + \
                 pd.DatetimeIndex(df['Date']).day

# Drop the original Date column
df.drop(columns=['Date'], inplace=True)

# Prepare features and target variable
x = df[['date_num']]  # Only using date_num for PCA
y = df['Close']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123456)

# Apply PCA on the date feature
pca = PCA(n_components=1)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)

# Train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train_pca, y_train)

# Save the PCA and model
with open('pca_and_model.pkl', 'wb') as file:
    pickle.dump((pca, model), file)

# Evaluate the model
y_train_pred = model.predict(x_train_pca)
y_test_pred = model.predict(x_test_pca)

# Mean Absolute Error
mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
print(f"Training MAE: {mae_train}")
print(f"Testing MAE: {mae_test}")

# Mean Squared Error
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
print(f"Training MSE: {mse_train}")
print(f"Testing MSE: {mse_test}")

# Root Mean Squared Error
rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)
print(f"Training RMSE: {rmse_train}")
print(f"Testing RMSE: {rmse_test}")

# R-squared score
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print(f"Training R2: {r2_train}")
print(f"Testing R2: {r2_test}")

# Load the PCA and model for prediction
with open('pca_and_model.pkl', 'rb') as file:
    pca, model = pickle.load(file)

# Given dates for prediction
given_date = pd.DataFrame({'date': ['2021-02-21', '2024-08-12', '2025-02-11', '2002-03-12']})
given_date['date_num'] = (pd.DatetimeIndex(given_date['date']).year % 100) * 10000 + \
                         pd.DatetimeIndex(given_date['date']).month * 100 + \
                         pd.DatetimeIndex(given_date['date']).day

# Prepare the input data for PCA transformation
input_data = given_date[['date_num']]
input_data_pca = pca.transform(input_data)  # Transform the input data using the fitted PCA

# Make the prediction using the model
y_pred = model.predict(input_data_pca)

# Print the predicted prices
print("Predicted prices:", y_pred)
```
